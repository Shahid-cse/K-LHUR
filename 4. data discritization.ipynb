## Discritization Fuction

```python
# Functions for data discretization

def discreteLength(length):
    if length < 100:
        return 0
    else:
        return 1

def discreteLane(lane):
    if lane < 2:
        return 0
    if lane < 3:
        return 1
    if lane < 4:
        return 2
    else:
        return 3

def discreteLevel(level):
    if level == 0 or level == 1:
        return 1
    if level == 2:
        return 2
    return 3

def discreteTor(tor):
    if tor < 1.25:
        return 0
    else:
        return 1
    
def discreteConnections(cnum):
    if cnum < 4:
        return 0
    if cnum < 5:
        return 1
    else:
        return 2

def discreteSumPOI(num):
    if num < 5:
        return 0
    return 1

def discretePOI(num):
    if num < 2:
        return 0
    if num < 4:
        return 1
    return 2

def discreteTime(timeSlot):
    if timeSlot < 7 * 6: #hour * hourly time slot num.
        return 0
    if timeSlot < 10 *6:
        return 1
    if timeSlot < 16 * 6:
        return 2
    if timeSlot < 20 * 6:
        return 3
    return 4

def discreteSup(support, level):
    if level == 1:
        if support < 3:
            return 0
        if support < 6:
            return 1
        if support < 10:
            return 2
        if support < 15:
            return 3
        if support < 25:
            return 4
        if support < 50:
            return 5
        return 6
    else:
        if level == 2:
            if support < 3:
                return 0
            if support < 6:
                return 1
            if support < 9:
                return 2
            if support < 12:
                return 3
            if support < 20:
                return 4
            if support < 40:
                return 5
            return 6
        else:
            if support < 1:
                return 0
            if support < 2:
                return 1
            if support < 3:
                return 2
            if support < 5:
                return 3
            if support < 10:
                return 4
            if support < 20:
                return 5
            return 6

# discretize the average speed
def discreteSpeed(speed, level):
    if level == 0 or level == 1:
        if speed < 10:
            return 0
        if speed < 15:
            return 1
        if speed < 17:
            return 2
        if speed < 19:
            return 3
        if speed < 21:
            return 4
        return 5
    else:
        if level == 2:
            if speed < 4:
                return 0
            if speed < 6:
                return 1
            if speed < 8:
                return 2
            if speed < 10:
                return 3
            if speed < 12:
                return 4
            return 5
        else:
            if speed < 2:
                return 0
            if speed < 3.5:
                return 1
            if speed < 5:
                return 2
            if speed < 6.5:
                return 3
            if speed < 8:
                return 4
            return 5
def discreteMaxSpeed(speed, level):
    if level == 0 or level == 1:
        if speed <= 5:
            return 0
        if speed < 10:
            return 1
        if speed <= 15:
            return 2
        if speed < 20:
            return 3
        if speed < 25:
            return 4
        return 5
    else:
        if level == 2:
            if speed < 4:
                return 0
            if speed < 6:
                return 1
            if speed < 8:
                return 2
            if speed < 10:
                return 3
            if speed < 12:
                return 4
            return 5
        else:
            if speed < 2:
                return 0
            if speed < 3.5:
                return 1
            if speed < 5:
                return 2
            if speed < 6.5:
                return 3
            if speed < 8:
                return 4
            return 5
def discreteStdSpeed(speed, level):
    if level == 0 or level == 1:
        if speed < 2:
             return 0
        if speed < 3.5:
            return 1
        if speed < 5:
            return 2
        if speed < 6.5:
            return 3
        if speed < 8:
            return 4
        return 5
    else:
        if level == 2:
            if speed < 4:
                return 0
            if speed < 6:
                return 1
            if speed < 8:
                return 2
            if speed < 10:
                return 3
            if speed < 12:
                return 4
            return 5
        else:
            if speed < 2:
                return 0
            if speed < 3.5:
                return 1
            if speed < 5:
                return 2
            if speed < 6.5:
                return 3
            if speed < 8:
                return 4
            return 5
```

```python
# This one for training data only
def discreteVolume(vol, level):
    if level == 1:
        if vol < 15:
            return 0
        if vol < 18.67:
            return 1
        if vol < 21.82:
            return 2
        if vol < 25.48:
            return 3
        return 4
    else:
        if level == 2:
            if vol < 3.03:
                return 0
            if vol < 6.19:
                return 1
            if vol < 8.91:
                return 2
            if vol < 12.07:
                return 3
            return 4
        else:
            if vol < 1.03:
                return 0
            if vol < 2.89:
                return 1
            if vol < 4.49:
                return 2
            if vol < 6.35:
                return 3
            return 4
```

```python
# Optimized Functions for data discretization

def discreteLength(length):
    return 1 if length < 100 else 2

def discreteLane(lane):
    return min(lane, 3)  # Lane values are capped at 3

def discreteLevel(level):
    return min(level, 3)  # Levels 1 -> 1; 2 -> 2; 3 and above -> 3

def discreteDirrection(dir):
    return (dir+1)

def discreteTor(tor):
    return 1 if tor < 1.25 else 2

def discreteConnections(cnum):
    return 1 if cnum < 4 else (2 if cnum < 5 else 3)

def discreteSumPOI(num):
    return 1 if num < 5 else 2

def discretePOI(num):
    if num < 2:
        return 1
    elif num < 4:
        return 2
    return 3

def discretize_weighted_POI(weighted_POI, level):
    if level == 1:
        # Level 1: Custom bins
        bins = [0, 1, 5, 10, 20, np.inf]  # 0 -> 1, 1-5 -> 2, 5-10 -> 3, 10-20 -> 4, >20 -> 5
    elif level == 2:
        # Level 2: Different bins based on data distribution
        bins = [0, 2, 8, 15, 25, np.inf]  # Adjusted bins for level 2
    elif level >= 3:
        # Level 3: Another set of bins
        bins = [0, 3, 10, 18, 30, np.inf]  # Adjusted bins for level 3
    
    # Use numpy's digitize to convert continuous values to discrete bins
    discretized_values = np.digitize(weighted_POI, bins)
    
    return discretized_values

def discreteTime(timeSlot):
    if timeSlot < 7 * 6:
        return 1
    elif timeSlot < 10 * 6:
        return 2
    elif timeSlot < 16 * 6:
        return 3
    elif timeSlot < 20 * 6:
        return 4
    return 5

def discreteSup(support, level):
    bins = [
        [3, 6, 10, 15, 25, 50],  # For level 1
        [3, 6, 9, 12, 20, 40],   # For level 2
        [1, 2, 3, 5, 10, 20]     # For level 3+
    ]
    thresholds = bins[min(level - 1, 2)]
    for i, threshold in enumerate(thresholds):
        if support < threshold:
            return i + 1
    return len(thresholds) + 1

def discreteSpeed(speed, level):
    bins = [
        [10, 15, 17, 19, 21],  # For level 0, 1
        [4, 6, 8, 10, 12],     # For level 2
        [2, 3.5, 5, 6.5, 8]    # For level 3+
    ]
    thresholds = bins[min(level-1, 2)]
    for i, threshold in enumerate(thresholds):
        if speed < threshold:
            return i + 1
    return len(thresholds) + 1

def discreteMaxSpeed(speed, level):
    if level in [0, 1]:
        bins = [5, 10, 15, 20, 25]
    elif level == 2:
        bins = [4, 6, 8, 10, 12]
    elif level >= 3:
        bins = [2, 3.5, 5, 6.5, 8]
    
    for i, threshold in enumerate(bins):
        if speed <= threshold:
            return i + 1
    return len(bins) + 1

def discreteStdSpeed(speed, level):
    bins = [
        [2, 3.5, 5, 6.5, 8],  # For level 0, 1
        [4, 6, 8, 10, 12],    # For level 2
        [2, 3.5, 5, 6.5, 8]   # For level 3+
    ]
    thresholds = bins[min(level-1, 2)]
    for i, threshold in enumerate(thresholds):
        if speed < threshold:
            return i + 1
    return len(thresholds) + 1
```
