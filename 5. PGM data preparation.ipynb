```python
# List of DataFrames for levels 1, 2, and 3
road_history_levels = {
    'Level 1': level1_road_history_holiday,
    'Level 2': level2_road_history_holiday,
    'Level 3': level3_road_history_holiday
}

# Iterating over each DataFrame
for level_name, df in road_history_levels.items():
    print(f"Processing {level_name}")
    print("=" * 60)  # Separator for readability
    
    # Getting unique values and their counts
    unique_values_counts = {col: (df[col].unique(), df[col].nunique()) for col in df.columns}
    
    # Printing the unique values and counts in a nicely formatted way
    for column, (unique_vals, count) in unique_values_counts.items():
        print(f"Column: {column}")
        print(f"Unique Values: {unique_vals}")
        print(f"Number of Unique Values: {count}")
        print("-" * 50)  # Separator for readability
    
    print("\n")  # Newline for separating different levels
```
