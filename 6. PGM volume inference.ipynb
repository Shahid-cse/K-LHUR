```python
output_dir = 'D:/Thesis/files_output_dir/output_preprocessing/'
# Load from a file
with open(output_dir+ 'vol_distribution_data.pkl', 'rb') as file:
    data = pickle.load(file)

vol_dict_level1 = data['vol_dict1']
vol_dict_level2 = data['vol_dict2']
vol_dict_level3 = data['vol_dict3']
mean_level1 = data['mean1']
mean_level2 = data['mean2']
mean_level3 = data['mean3']


# Define the function with vol_dict as an additional parameter
def map_class_to_volume(cls, vol_dict):
    low, high = vol_dict[cls+1]
    return (low + high) / 2  # Use the midpoint of the range as the most probable value
```

```python
# Apply the function using .map() and a lambda to pass vol_dict
level1_road_history_workday['volume'] = level1_road_history_workday['volume_claas'].map(lambda cls: map_class_to_volume(cls, vol_dict_level1))
level1_road_history_holiday['volume'] = level1_road_history_holiday['volume_claas'].map(lambda cls: map_class_to_volume(cls, vol_dict_level1))

level1_road_history_workday['volume'] = level1_road_history_workday['volume'] * level1_road_history_workday['lanes']
level1_road_history_holiday['volume'] = level1_road_history_holiday['volume'] * level1_road_history_holiday['lanes']
```
